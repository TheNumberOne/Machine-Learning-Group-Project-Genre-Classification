```python
import pandas as pd
from extract_words import extract_words
from collections import Counter
```

```python

data = pd.read_csv('gutenberg.csv', index_col='gutenberg id')
data
```

```python
words = Counter()
for gutenberg_id in data.index:
    words.update(extract_words(gutenberg_id))
```
