```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
books = pd.read_csv("997-books-dataset-no-empties.csv")
books
```

```python
subject_counts = Counter(books.genre)
subject_freqs = pd.DataFrame(data=subject_counts.most_common(5), columns=["subject", "freq"])
subject_freqs.to_csv("5-most-common-subjects.csv")
```

```python
most_common_subjects = { subject: freq for subject, freq in subject_counts.most_common() if freq >= 10 }
books_with_common_subjects = books[books.genre.isin(most_common_subjects)]
books_with_common_subjects
```

```python
books_with_common_subjects.to_csv(f"{len(books_with_common_subjects)}-books-with-common-subjects.csv")
```
